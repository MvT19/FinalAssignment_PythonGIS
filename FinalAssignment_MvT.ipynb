#### Assignment 1: Download a shapefile of the world from [naturalearthdata.com](https://www.naturalearthdata.com) and perform a [left join](https://chrisalbon.com/python/data_wrangling/pandas_join_merge_dataframe/) with the attached file "under5mortality.csv", which contains child mortality rates for each country in the world. The join should be done on the ISO_A3 column in the shapefile and the ISO column in the CSV, respectively.

First import the libraries that are needed:

```python
import wget
import geopandas as gpd 
import pandas as pd
from pysal.lib import weights
from pysal.explore import esda
```

Download the shapefile

```python
url="https://www.naturalearthdata.com/http//www.naturalearthdata.com/download/10m/cultural/ne_10m_admin_0_countries.zip"
filename = wget.download(url)
print ('filename is ', filename)
zip = zipfile.ZipFile(filename)
zip.extractall('.')
zip.close()
```

```python
curwd= os.chdir("D:/Documents/Python") #Set filepath 
fp = "ne_10m_admin_0_countries.shp" 

# Read file using gpd.read_file() 
data = gpd.read_file(fp) #ead in as a geodataframe

print ("The type of data is", type(data))
```

```python
mortality= pd.read_csv("under5mortality.csv")
mortality
```

Perform the left_join on "ISO_A3" and "ISO"

```python
world_mortality = pd.merge(data, mortality, left_on="ISO_A3",right_on="ISO", how="left")
```

#### Assignment 2: Calculate the spatial weights matrix for the world based on border neighborhood and use the weights to calculate Moran's I for child mortality.

First make sure that child mortality does not include Na's

```python
pd.isnull(world_mortality['ChildMortality']).sum() # The column has 62 Na's
```

Exclude the rows with Na's from further calculations

```python
child_mortality = world_mortality[world_mortality['ChildMortality'].isnull() == False]
child_mortality = child_mortality.set_index(np.arange(child_mortality.shape[0]))
```

```python
pd.isnull(child_mortality['ChildMortality']).sum() # check for Na's again
```

Calculate the spatial weights matrix. Since *border neighborhood* can be interpreted differently, I compared the matrices based on the Queen's (shared edges and vertices) and the Rook's case (shared edges).

```python
w_rook = weights.Rook.from_dataframe(child_mortality)
w_queen = weights.Queen.from_dataframe(child_mortality) 
```

Plot matrices for botht the Queen's and the Rook's case

```python
f, axs = plt.subplots(1, 2, figsize=(16, 16))

child_mortality.plot(color='k', ax=axs[0], 
        edgecolor='w', linewidth=0.5)
w_rook.plot(child_mortality, ax=axs[0], color='red')
axs[0].set_title('Rook weights')

child_mortality.plot(color='k', ax=axs[1], 
        edgecolor='w', linewidth=0.5)
w_queen.plot(child_mortality, ax=axs[1], color='red')
axs[1].set_title('Queen weights')

plt.show()
```

Calculate the Moran's I for child mortality, using both weight matrices

```python
moran_queen = esda.Moran(child_mortality['ChildMortality'], w_queen)
moran_rook = esda.Moran(child_mortality['ChildMortality'], w_rook)
```

```python
print("Moran's I with the Queen's case: ",moran_queen.I)
print("Moran's I with the Rook's case: ",moran_rook.I)
```

```python
print(moran_queen.p_sim, "<0.05: signification spatial correlation")
print(moran_rook.p_sim, "<0.05: signification spatial correlation")
```
